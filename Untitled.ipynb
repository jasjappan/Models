```python
class DoubleConv(nn.Module):
  def __init__(self, in_channels, out_channels):
      super(DoubleConv, self).__init__()
      self.conv1 = nn.Conv2d(in_channels,out_channels,kernel_size=3,padding=1)
      self.norm = nn.BatchNorm2d(out_channels)
      self.relu = nn.ReLU(inplace=True)
      self.conv2 = nn.Conv2d(out_channels,out_channels,kernel_size=3,padding=1)
      self.norm2 = nn.BatchNorm2d(out_channels)
      self.out = nn.ReLU(inplace=True)

  def forward(self,x):
      x = self.conv1(x)
      x = self.norm(x)
      x = self.relu(x)
      x = self.conv2(x)
      x = self.norm2(x)
      x = self.out(x)
      return x

class UpBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super(UpBlock, self).__init__()
        self.up = nn.ConvTranspose2d(in_channels, out_channels, kernel_size=2, stride=2)
        # After concatenation: out_channels (from upsampling) + out_channels (from skip) = 2*out_channels
        self.conv = DoubleConv(out_channels * 2, out_channels)

    def forward(self, x1, x2):
        x1 = self.up(x1)  # Upsample x1

        # Handle size differences
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]
        x1 = TF.pad(x1, [diffX // 2, diffX - diffX // 2,
                        diffY // 2, diffY - diffY // 2])

        # Concatenate skip connection (x2) with upsampled feature (x1)
        x = torch.cat([x2, x1], dim=1)
        x = self.conv(x)
        return x


class Unet(nn.Module):
    def __init__(self, in_channels, out_channels,features = [32,64,128,256,512]):
        super(Unet,self).__init__()
        self.enc1  = DoubleConv(in_channels,features[0])
        self.Mp1 = nn.MaxPool2d(2)
        self.enc2 = DoubleConv(features[0],features[1])
        self.Mp2 = nn.MaxPool2d(2)
        self.enc3 = DoubleConv(features[1],features[2])
        self.Mp3 = nn.MaxPool2d(2)
        self.enc4 = DoubleConv(features[2],features[3])
        self.Mp4 = nn.MaxPool2d(2)
        self.bottleNeck = DoubleConv(features[3],features[4])
        self.upsample  = UpBlock(features[4],features[3])
        self.upsample2 = UpBlock(features[3],features[2])
        self.upsample3 = UpBlock(features[2],features[1])
        self.upsample4 = UpBlock(features[1],features[0])
        self.final_conv = nn.Conv2d(features[0], out_channels, kernel_size=1)

    def forward(self,x):
        x = self.enc1(x)
        x1 = x
        x = self.Mp1(x)
        x = self.enc2(x)
        x2 = x
        x = self.Mp2(x)
        x = self.enc3(x)
        x3 = x
        x = self.Mp3(x)
        x = self.enc4(x)
        x4 = x
        x = self.Mp4(x)
        x = self.bottleNeck(x)
        x = self.upsample(x,x4)
        x = self.upsample2(x,x3)
        x = self.upsample3(x,x2)
        x = self.upsample4(x,x1)
        x = self.final_conv(x)

        return x
```

```python
class Cityscape(Dataset):
    def __init__(self, dcit, transform=None, label_map='id'):
        self.transform = transform
        self.ds = dcit
        self.label_map = label_map
        self.id_2_train = {}
        self.id_2_cat = {}
        self.trainid_2_name = {}
        self.train_2_id = {}
        self.id_2_name = {}
        self.color_2_id = {}  # RGB color to ID mapping
        self.color_2_trainid = {}  # RGB color to trainID mapping
        self.color_2_catid = {}  # RGB color to categoryID mapping
        
        # Build mapping dictionaries
        for lbl in labels:
            self.id_2_train[lbl.id] = lbl.trainId
            self.id_2_cat[lbl.id] = lbl.categoryId
            self.color_2_id[lbl.color] = lbl.id
            self.color_2_trainid[lbl.color] = lbl.trainId
            self.color_2_catid[lbl.color] = lbl.categoryId
            
            if lbl.trainId != 255:  # Valid trainId
                self.trainid_2_name[lbl.trainId] = lbl.name
                self.train_2_id[lbl.trainId] = lbl.id
            if lbl.id > 0:
                self.id_2_name[lbl.id] = lbl.name

    def rgb_to_mask(self, rgb_mask):
        """Convert RGB mask to label mask based on label_map setting"""
        h, w, c = rgb_mask.shape
        mask = np.zeros((h, w), dtype=np.int16)  # Use int16 to handle -1 values
        
        if self.label_map == 'id':
            color_map = self.color_2_id
        elif self.label_map == 'trainId':
            color_map = self.color_2_trainid
        elif self.label_map == 'categoryId':
            color_map = self.color_2_catid
        else:
            raise ValueError(f"Unknown label_map: {self.label_map}")
        
        # Map each RGB color to corresponding label
        for color, label_value in color_map.items():
            # Find pixels matching this color
            matches = np.all(rgb_mask == color, axis=2)
            if label_value == -1:  # Handle license plate case
                if self.label_map == 'id':
                    mask[matches] = 0  # Map to unlabeled
                else:
                    mask[matches] = 255  # Map to ignore index
            else:
                mask[matches] = label_value
        
        # Convert back to appropriate dtype
        if self.label_map == 'id':
            # Handle any remaining -1 values and convert to uint8
            mask[mask == -1] = 0
            mask = mask.astype(np.uint8)
        elif self.label_map == 'trainId':
            # trainId uses 255 as ignore index, so uint8 is fine
            mask = mask.astype(np.uint8)
        else:
            mask = mask.astype(np.uint8)
            
        return mask

    def __getitem__(self, idx):
        label_dict = self.ds['train'][idx]['imageA']
        mask_dict = self.ds['train'][idx]['imageB']
        
        # Load image
        image = Image.open(BytesIO(label_dict['bytes'])).convert("RGB")
        
        # Load mask as RGB and convert to numpy
        mask_img = Image.open(BytesIO(mask_dict['bytes'])).convert("RGB")
        rgb_mask = np.array(mask_img).astype(np.uint8)
        
        # Convert RGB mask to label mask
        mask = self.rgb_to_mask(rgb_mask)
        
        # Handle invalid labels
        if self.label_map == 'trainId':
            # Set invalid trainIds (255) to ignore index
            mask[mask == 255] = 255  # Keep as ignore index
        elif self.label_map == 'id':
            # Handle negative IDs (like license plate with id=-1)
            mask[mask == 255] = 0  # Map to unlabeled
            
        # Apply transforms
        if self.transform:
            image = self.transform(image)
            # Note: Be careful with mask transforms - they should preserve label values
            # Only apply transforms that don't change pixel values (like ToTensor)
            if hasattr(self.transform, 'transforms'):
                # If using Compose, apply only image transforms to image
                # and only tensor conversion to mask
                pass
        
        # Convert mask to tensor
        mask = torch.from_numpy(mask).long()
        
        return image, mask
            
    def __len__(self):
        return len(self.ds['train'])

    def get_class_names(self):
        """Return list of class names for current label mapping"""
        if self.label_map == 'trainId':
            return [self.trainid_2_name.get(i, f'class_{i}') for i in range(19)]
        elif self.label_map == 'id':
            return [self.id_2_name.get(i, f'class_{i}') for i in range(34)]
        else:
            return []
            
    def get_num_classes(self):
        """Return number of classes for current label mapping"""
        if self.label_map == 'trainId':
            return 19  # Standard Cityscapes has 19 trainable classes
        elif self.label_map == 'id':
            return 34  # Total number of defined IDs
        elif self.label_map == 'categoryId':
            return 8   # Number of categories
        return 0
```

```python
epochs=100
verbose =1
model.train()
for epoch in range(epochs):
        running_loss = 0.0
        total_batches = len(data_loader)

        print(f"\nEpoch {epoch+1}/{epochs}")
        print("-" * 30)

        # tqdm progress bar over batches
        progress_bar = tqdm(enumerate(data_loader), total=total_batches, leave=False)
        for batch_idx, (images, masks) in progress_bar:
            images = images.to(device)
            masks = masks.to(device)

            optimizer.zero_grad()
            outputs = model(images)

            loss = criterion(outputs, masks)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            avg_loss = running_loss / (batch_idx + 1)
            progress_bar.set_description(f"Batch {batch_idx+1}/{total_batches}")
            progress_bar.set_postfix(loss=f"{avg_loss:.4f}")

        epoch_loss = running_loss / total_batches
        print(f"Epoch {epoch+1} finished. Average Loss: {epoch_loss:.4f}")
```

```python
def inference(model, dataloader, device):
    model.to(device)
    model.eval()
    
    preds = []
    with torch.no_grad():
        for images, _ in dataloader:  # Ignore labels during inference
            images = images.to(device)  # (N, 3, H, W)
            
            outputs = model(images)      # (N, 35, H, W)
            
            # Get predicted class per pixel (argmax over class dim)
            predicted_masks = torch.argmax(outputs, dim=1)  # (N, H, W)
            
            preds.append(predicted_masks.cpu())
    
    # Concatenate all predictions
    preds = torch.cat(preds, dim=0)
    return preds
```
